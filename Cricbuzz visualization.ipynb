```python
df.groupby(['Type'], as_index = False)['Runs'].sum().sort_values(by='Runs', ascending = False)
```

```python
High_total = df.groupby(['Type'], as_index = False)['Runs'].sum().sort_values(by='Runs', ascending = False)

sns.barplot(x = 'Type', y = 'Runs', data = High_total)
```

```python
T_R = df.groupby(['Player'], as_index = False)['100'].sum().sort_values(by = '100', ascending = False).head(10)

plt.figure(figsize =(15,5))
sns.barplot(data = T_R, x = 'Player', y = '100')
```

```python
top_10_players = df.sort_values(by='Runs', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_10_players['Player'], top_10_players['Runs'], color='purple')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.title('Top 10 Players by Runs')

plt.show()
```

```python
best = df.groupby(['Player'], as_index = False)['Avg'].sum().sort_values(['Avg'], ascending = True).head(15)
plt.figure(figsize = (25,5)) 
sns.barplot(data = best, x = 'Player', y = 'Avg')
```

```python
subset_data = df.sample(n=100) 

# Create a color-coded scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runs', y='Avg', hue='SR', data=subset_data, palette='viridis', size='SR', sizes=(10, 200))
plt.xlabel('Runs')
plt.ylabel('Batting Average (Avg)')
plt.title('Relationship between Runs, Batting Average, and Strike Rate')
plt.legend(title='Strike Rate (SR)')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.bar(df, x='Player', y='Runs', title='Runs Scored by Each Player')
fig.show()
```

```python
numeric_columns = ['Runs', 'HS', 'Avg', 'BF', 'SR', '4s', '6s', '50', '100']

# Calculate the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.title("Correlation Heatmap of Cricket Player Statistics")
plt.show()
```
